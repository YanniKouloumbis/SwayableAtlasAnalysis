```python
user_id = '6196e8a92545cb00273706dc'

map = project.maps[0]
original_vector = project.get_data(ids=[user_id])
print("ORIGINAL VECTOR: ", original_vector)

neighbors, distances = map.vector_search(ids=[user_id], k=5)

print("NEIGHBORS: ", neighbors)
print("DISTANCES: ", distances)
```

```python
neighbor_data = project.get_data(ids=neighbors[0])
print(f"The five nearest neighbors to the query point are:")
for i, neighbor in enumerate(neighbor_data):
    print(f"Neighbor {i+1}: {neighbor['answers']}")
```

Generative AI: Given the data of the 5 neighbors, use Generative AI to construct a user sentiment profile.

```python
import openai
import os

openai.api_key = os.getenv("OPENAI_KEY")

prompt = f"""Below are 6 users who submitted an answer about watching a specific video:
{original_vector}
{neighbor_data}
Based on their user characteristics and shared traits among their neighbours, come up with the underlying reasons of why they liked or disliked this particular ad. 
For example, when a user says:

I think the features of the account are great and could fill a big need for some people. It’s a good advertisement

And their particular profile is: 

ID: 274
hash: 6196e8a92545cb00273706dc
treatment_group: Chime - Banking with No Hidden Fees
Age: 35-54: 0
Age: 55+: 1
Age: Under 35: 0
Annual Income: $50-99K: 0
Annual Income: HHI $100K or more: 1
Annual Income: Less than $50K: 0
Banking Hidden Fee Experience: No, never: 0
Banking Hidden Fee Experience: Yes, many times: 1
Banking Hidden Fee Experience: Yes, once or twice: 0
Current Bank: Banking with Chime: 1
Current Bank: Currently with other bank: 0
Current Bank: Unbanked: 0
Education: College: 1
Education: Non-College: 0
Education: Postgrad: 0
Ethnicity: Asian: 0
Ethnicity: Black: 0
Ethnicity: Latinx: 0
Ethnicity: Other: 0
Ethnicity: White: 1
Gender: Female: 1
Gender: Male: 0
Region: Midwest: 0
Region: Northeast: 0
Region: South: 1
Region: West: 0
Social Media Use: TikTok Users: 0
Urbanization: Rural: 0
Urbanization: Suburban: 1
Urbanization: Urban: 0
isReal: 1
Please share any reactions or feedback you have on the video: I think the features of the account are great and could fill a big need for some people. It’s a good advertisement

This could be summarized as:

User 274 is a 55+ year old woman who lives in the South. She enjoys a comfortable life with an annual income of $100K or more and resides in a suburban neighborhood. She is a college-educated individual with a strong preference for Chime, her current banking service.

In the past, she has experienced hidden banking fees many times, which might be a reason for her appreciation of Chime's no-hidden-fees approach. Her ethnicity is White, and she does not use TikTok for social media purposes.

This user values Chime's features, believing that they fill a significant need for many people. She considers the advertisement promoting Chime's account to be effective and well-executed.

Underlying reason: User 274 likes the Chime ad because it emphasizes no-hidden-fees, addressing her past negative experiences and valuing transparency.
"""


response = openai.ChatCompletion.create(
      model="gpt-3.5-turbo",
      messages=[{
        'role': 'system',
        'content': "You are a user researcher at Swayable, and you are looking at the responses of users who watched a specific video. You want to understand why the users responded the way they did."
        }, {
        'role': 'user',
        'content': prompt,
        }],
        max_tokens=1024,
        temperature=0.9,
)

print(response['choices'][0]['message']['content'])
```
